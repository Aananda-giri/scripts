# use yt_dlp code 
note: no sound in video
* take url as input
* give option to select format to download
* download the  video or audio based on format selected by user

```python
import yt_dlp
import os

def download_video_audio(url):
    # Create an instance of yt_dlp
    ydl = yt_dlp.YoutubeDL()

    # Get available formats for the given URL
    info = ydl.extract_info(url, download=False)
    formats = info['formats']

    # Display available formats to the user
    print("Available Formats:")
    for i, format in enumerate(formats):
        print(f"{i+1}. {format['ext']} - {format['format_note']}")

    # Prompt user to select a format
    format_index = int(input("Enter the format number to download: ")) - 1
    selected_format = formats[format_index]

    # Set the output directory for the downloaded file
    output_dir = os.getcwd()  # Current working directory

    # Set the output file path
    output_file = os.path.join(output_dir, f"{info['title']}.{selected_format['ext']}")

    # Prepare options for download
    options = {
        'format': selected_format['format_id'],
        'outtmpl': output_file,
    }

    # Download the video or audio
    with yt_dlp.YoutubeDL(options) as ydl:
        ydl.download([url])

    print("Download complete!")

# Example usage
# url = input("Enter the YouTube video URL: ")
url = 'https://youtu.be/crlIG8QgyiQ'
download_video_audio(url)
```
